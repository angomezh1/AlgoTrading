```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker
import matplotlib.font_manager as font_manager
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.metrics import median_absolute_error
from matplotlib.gridspec import  GridSpec
from pandas.plotting import scatter_matrix
from scipy.optimize import curve_fit


df = pd.read_csv('D:/andrgome/Documents/BDPython/BD_ProbdefaultColombia.csv',encoding='latin-1')
df_cds = pd.read_csv('D:/andrgome/Documents/BDPython/BDCDS5Y.csv',encoding='latin-1')
df_countries = pd.read_csv('D:/andrgome/Documents/BDPython/BD_CDS_COUNTRY.csv',encoding='latin-1')
df_countries.sort_values(by=['Actual Credit Rating'],inplace=True)


#X = df[['CTACTE','DEUDAEXT','DEUDAPIB','BALFISCAL','INTPIB','IMPTOSPIB','RINPIB','PFIPIB','GSFC','EMBI']]
X = df[['CTACTE','DEUDAPIB','INTPIB','IMPTOSPIB','EMBI']]
y = df['PD']


#SI SE QUIERE MODIFICAR EL PRONÓSTICO DEL ÚLTIMO DATO
# Z=[-0.045,0.55,0.03,0.135,350]
# X.iloc[-1] = Z

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=3)

# linearM  = LinearRegression(normalize=True)
# linearM.fit(X_train,y_train)
# print('wi OLS =',linearM.coef_)
# print('w0 OLS =',linearM.intercept_)
# print('R2 OLS =',linearM.score(X_test,y_test))

ridge  = Ridge(alpha=1,normalize=True)
ridge.fit(X_train,y_train)
print('wi Ridge =',ridge.coef_)
print('w0 Ridge =',ridge.intercept_)
print('R2 Ridge =',ridge.score(X_test,y_test))

# lasso  = Lasso(alpha=10,normalize=True)
# lasso.fit(X_train,y_train)
# print('wi Lasso =',lasso.coef_)
# print('w0 Lasso =',lasso.intercept_)
# print('R2 Lasso =',lasso.score(X_test,y_test))

y_pred = ridge.predict(X)

print('MAE =',median_absolute_error(y, y_pred))
MAE = median_absolute_error(y, y_pred)

banda_superior = y_pred + 2*MAE
banda_inferior = y_pred - 2*MAE

#Gráfico
font = font_manager.FontProperties(family='arial',weight='bold', style='normal', size=20)
tick_spacing1 = 5
tick_spacing2 = 50

fig = plt.figure()

ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(df['Fecha'],y_pred,marker='o', linestyle='solid',linewidth=5,color='goldenrod',alpha=0.99,label='MODELO',markersize=1)
ax1.plot(df['Fecha'],df['PD'],marker='o',linestyle='solid',linewidth=5,color='darkblue',alpha=0.99,label='PROB INCUMP',markersize=1)
ax1.fill_between(df['Fecha'],banda_inferior,banda_superior,color='lightgrey', alpha= 0.7)
ax1.legend(frameon=False,prop=font,loc=1, bbox_to_anchor=(0.9, 1),ncol=10,markerscale=10)
ax1.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing1))
plt.xticks(rotation=90)
plt.yticks(fontname = "Arial",fontsize=20)
plt.xticks(fontname = "Arial",fontsize=20)
ax1.title.set_fontfamily('arial')
ax1.axis([None, None, None, None])
vals = ax1.get_yticks()
ax1.set_yticklabels(['{:,.0%}'.format(x) for x in vals])
#plt.box(on=None)

# ax2 = ax1.twinx()
# ax2.plot(df['Fecha'],X.values.dot(ridge.coef_),marker='o',linestyle='solid',linewidth=5,color='goldenrod',alpha=0.99,label='Z-Score Macro',markersize=1)
# ax2.legend(frameon=False,prop=font,loc=1, bbox_to_anchor=(1, 1),ncol=10,markerscale=3)
# ax2.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing2))
# plt.yticks(fontname = "Arial",fontsize=25)
# plt.xticks(fontname = "Arial",fontsize=25)
# #ax2.axis(['2018-12-31', '2020-03-20', None, None])
# vals = ax2.get_yticks()
# ax2.set_yticklabels(['{:,.2}'.format(x) for x in vals])

plt.box(on=None)
plt.subplots_adjust(bottom=0.1, right=5.4, top=1.9)
plt.show()


#GRÁFICOS PARA RELACIONAR EL CDS CON LA PROBABILIDAD DE DEFAULT
# ax3 = fig.add_subplot(1, 2, 2)
# ax3.plot(df_cds['Fecha'],df_cds['CDS5Y'],marker='None',linestyle='solid',linewidth=5,color='darkblue',alpha=0.99,label='CDS 5Y COL',markersize=10)
# ax3.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing2 ))
# ax3.legend(frameon=False,prop=font,loc=1, bbox_to_anchor=(0.5, 1),ncol=10,markerscale=10)
# plt.xticks(rotation=90)
# plt.yticks(fontname = "Arial",fontsize=20)
# plt.xticks(fontname = "Arial",fontsize=20)
# ax3.title.set_fontfamily('arial')
# ax3.axis([None, None, None, None])
# vals = ax3.get_yticks()
# ax3.set_yticklabels(['{:,.0f}'.format(x) for x in vals])
# plt.box(on=None)

# ax4 = ax3.twinx()
# ax4.plot(df_cds['Fecha'],df_cds['PD'],marker='o',linestyle='solid',linewidth=5,color='goldenrod',alpha=0.99,label='PROB INCUMP',markersize=1)
# ax4.legend(frameon=False,prop=font,loc=1, bbox_to_anchor=(1, 1),ncol=10,markerscale=3)
# ax4.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing2))
# plt.yticks(fontname = "Arial",fontsize=20)
# plt.xticks(fontname = "Arial",fontsize=20)
# ax4.axis([None, None, None, None])
# vals = ax4.get_yticks()
# ax4.set_yticklabels(['{:,.0%}'.format(x) for x in vals])
# plt.box(on=None)



#GRÁFICOS PARA MIRAR LA EVOLUCIÓN DE LAS VARIABLES MACRO

# font = font_manager.FontProperties(family='arial',weight='bold', style='normal', size=35)
# tick_spacing1 = 5
# tick_spacing2 = 50

# ax3 = fig.add_subplot(1, 2, 1)
# ax3.plot(df['Fecha'],df['CTACTE'],marker='None',linestyle='solid',linewidth=10,color='darkblue',alpha=0.99,label='CTA CTE',markersize=10)
# ax3.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing2 ))
# ax3.legend(frameon=False,prop=font,loc=1, bbox_to_anchor=(0.5, 1.2),ncol=10,markerscale=10)
# plt.xticks(rotation=90)
# plt.yticks(fontname = "Arial",fontsize=35)
# plt.xticks(fontname = "Arial",fontsize=35)
# ax3.title.set_fontfamily('arial')
# ax3.axis([None, None, None, None])
# vals = ax3.get_yticks()
# ax3.set_yticklabels(['{:,.2%}'.format(x) for x in vals])
# plt.box(on=None)

# ax4 = ax3.twinx()
# ax4.plot(df['Fecha'],df['DEUDAPIB'],marker='o',linestyle='solid',linewidth=10,color='goldenrod',alpha=0.99,label='DEUDA-PIB (eje der)',markersize=1)
# ax4.legend(frameon=False,prop=font,loc=1, bbox_to_anchor=(1.0, 1.2),ncol=10,markerscale=3)
# ax4.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing1))
# plt.yticks(fontname = "Arial",fontsize=35)
# plt.xticks(fontname = "Arial",fontsize=35)
# ax4.axis([None, None, None, None])
# vals = ax4.get_yticks()
# ax4.set_yticklabels(['{:,.0%}'.format(x) for x in vals])
# plt.box(on=None)


# ax5 = fig.add_subplot(1, 2, 2)
# ax5.plot(df['Fecha'],df['INTPIB'],marker='None',linestyle='solid',linewidth=10,color='darkblue',alpha=0.99,label='INT-PIB',markersize=10)
# ax5.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing1 ))
# ax5.legend(frameon=False,prop=font,loc=1, bbox_to_anchor=(0.5, 1.2),ncol=10,markerscale=10)
# plt.xticks(rotation=90)
# plt.yticks(fontname = "Arial",fontsize=35)
# plt.xticks(fontname = "Arial",fontsize=35)
# ax4.title.set_fontfamily('arial')
# ax4.axis([None, None, None, None])
# vals = ax5.get_yticks()
# ax5.set_yticklabels(['{:,.2%}'.format(x) for x in vals])
# plt.box(on=None)

# ax6 = ax5.twinx()
# ax6.plot(df['Fecha'],df['IMPTOSPIB'],marker='o',linestyle='solid',linewidth=10,color='goldenrod',alpha=0.99,label='IMPTOSPIB (eje der)',markersize=1)
# ax6.legend(frameon=False,prop=font,loc=1, bbox_to_anchor=(1.0, 1.2),ncol=10,markerscale=3)
# ax6.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing1))
# plt.yticks(fontname = "Arial",fontsize=35)
# plt.xticks(fontname = "Arial",fontsize=35)
# ax6.axis([None, None, None, None])
# vals = ax6.get_yticks()
# ax6.set_yticklabels(['{:,.0%}'.format(x) for x in vals])
# plt.box(on=None)
# plt.subplots_adjust(bottom=0.1, right=5.4, top=1.9)
# plt.show()




```
